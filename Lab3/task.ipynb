# Вроцлавская таксономия: метод дендрита


**Вроцлавская таксономия** — метод многомерной классификации, при котором объекты группируются на основе минимальных различий между ними.

Метод также называется **методом дендритов**.


![вроц](vroc.png)

**Дендрит** — это граф без циклов (то есть ломаная линия), который:

- может ветвиться,
- соединяет любые две вершины,
- не содержит замкнутых путей.

Дендрит используется для нелинейного упорядочивания объектов, что позволяет выявить сложные взаимосвязи между ними.



![Дендрит](dendrite.png)

Задача метода — построить такой дендрит, в котором:

- смежные объекты имеют минимальные различия по признакам;
- суммарная длина связей между объектами минимальна.

Это позволяет получить оптимальную структуру упорядочивания.


**Этапы построения дендрита**
1. Построить матрицу расстояний между всеми объектами.
2. Для каждой строки (объекта) найти ближайший объект (наименьшее расстояние).
3. Построить связи между ближайшими объектами.
4. Объединить полученные пары в кластеры 1-го порядка.
5. Объединять кластеры, находя минимальные расстояния между ними, пока не получится единая структура.


Метод может быть использован в рамках классификации задач по Вроцлавской таксономии:

- **Р** — репродукция (воспроизведение знаний),
- **М** — модификация (адаптация и улучшение),
- **СК** — синтез компонентов (составление новой структуры),
- **НЗ** — новая задача (полная новизна, эвристика).

Тип задачи зависит от того, насколько творчески используется информация при построении дендрита.


Метод дендритов широко применяется:

- в экономике (группировка регионов или предприятий),
- в социологии (типология поведения),
- в биологии (кластеризация видов),
- в психологии (типология личности).

Преимущество метода — выявление сложных, неиерархических структур между объектами.
